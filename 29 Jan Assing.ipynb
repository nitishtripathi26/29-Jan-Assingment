```python
Q5-Is python code compiled or interrupted?

Ans-Python code is typically compiled and executed in a single step, a process known as "interpreted" execution.

When you run a Python program, the interpreter reads the source code line by line, parses it, compiles it into
bytecode, and executes the bytecode immediately. This is in contrast to compiled languages like C or Java, where
the code is compiled into machine code before it can be executed.

So while Python is not strictly a compiled language in the traditional sense, it does use a compilation step as
part of its interpretation process to improve performance.
```

```python
Q9-Python supports the creation of anoymous function at runtime, using a construct called----------

Ans-Python supports the creation of anonymous functions at runtime, using a construct called "lambda" functions.

Lambda functions are a way to define small, one-line functions without giving them a name. They can be used in
place of a regular function anywhere a function is expected, such as as an argument to a higher-order
function like 'map()' or 'filter()', or as a key function for sorting

The syntax for a lambda function is as follows:

   lambda arguments: expression

Here, arguments is a comma-separated list of 'arguments' that the function will take, and 'expression' is a single
expression that will be evaluated and returned by the function. The result is a function object that can be
called like any other function.

```

```python
Q14-How is memory managed in python?

Ans- Memory management in Python is handled by the Python memory manager. The memory manager is responsible 
for allocating and releasing memory for the objects created in Python. Python uses automatic memory management,
which means that the memory manager takes care of allocating memory for objects when they are created, and
releasing memory when they are no longer needed.

Python uses a technique called reference counting to manage memory. When an object is created, the memory 
manager allocates memory for the object and assigns a reference count of 1 to it. The reference count is
increased each time the object is referred to by another object or variable. When an object is no
longer needed, the reference count is decreased, and if it reaches 0, the memory occupied by the
object is released.

Overall, 'Python's memory management system is designed to be efficient and reliable, while providing automatic
memory management that makes it easier to write and maintain Python code.
```
